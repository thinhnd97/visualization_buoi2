```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'D:\Data\HUS_Master\Visualization\nigeria_houses_data.csv')
df = df.loc[df.state.isin(['Lagos', 'Abuja', 'Ogun', 'Oyo', 'Rivers', 'Imo', 'Anambara', 'Enugu', 'Edo'])]
df = df.loc[(df.price < 5e9) & (df.price > 1e6)]
df['price_log'] = np.log(df['price'])
```

```python
df.state.value_counts()
```

```python
correlation_mat = df.corr()
sns.heatmap(correlation_mat, annot=True)
```

```python
df.boxplot(column=['price_log'], by='state', figsize=(12,8), grid=False, vert=False)
```

```python
plt.figure(facecolor='w')
fig = plt.figure(figsize=(12, 10), facecolor='w')
i = 0
for state in df['state'].unique():
    i += 1
    data = df.loc[df.state == state]
    plt.subplot(3, 3, i)
    plt.scatter(data.bedrooms, data.price_log)
    plt.title(state)
fig.suptitle("BEDROOMS - PRICE", fontsize=30, fontfamily='serif', color='darkred', weight='bold')
```
